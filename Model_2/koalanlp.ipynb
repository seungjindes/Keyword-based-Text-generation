```python
import koalanlp
from koalanlp import API
from koalanlp.proc import Parser
from koalanlp.Util import initialize, finalize, PhraseTag


def find_obj (sentence) :

    initialize(java_options="-Xmx4g", HNN="LATEST" , KKMA = "LATEST")
    parser = Parser(API.KKMA)

    parsed = parser(sentence)
    obj = []
    
    for dep in parsed[0].getDependencies() :
        print(dep)
        depType = dep.getType()
        if (depType == PhraseTag.NP) :
          dep0 = dep.getDest()
          key = dep0.singleLineString().split('/')[0]
          obj.append(key)
          
    print()
    
    finalize()
    return obj


print(find_obj("늘 회사에서 갑자기 전화가 와서 너무 짜증이 났는데 너한테 이야기 하고 나니까 마음이 좀 풀렸어 앞으로도 내가 힘들 때 내 이야기 좀 들어 줘"))
```
